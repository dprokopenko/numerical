## Численное решение дифференциальных уравнений
* Метод Эйлера (частный случай формулы Рунге-Кутта)
$$ y'=f(x,y) $$
$$ y_{i+1}=y_i+h \cdot f(x_i,y_i) $$
$$ h=\frac{b-a}{n} $$

* Формула Рунге-Кутта
$$ y'=f(x,y) $$
$$ y_{i+1}=y_i+\frac{h}{6} (k_1+2k_2+2k_3+k_4) $$
$$ k_1=f(x_i,y_i) \cdot h $$
$$ k_2=f(x_i+\frac{h}{2},y_i+\frac{k_1}{2}) \cdot h $$
$$ k_3=f(x_i+\frac{h}{2},y_i+\frac{k_2}{2}) \cdot h $$
$$ k_4=f(x_i+h,y_i+k_3) \cdot h \\ h=\frac{b-a}{n} $$
Мы воспользуемся
$$ y_{i+1}=y_i+f(x_i+\frac{h}{2},y_i+\frac{k_1}{2}) \cdot h $$

```python
import diffeqs
from matplotlib import pyplot as plt
```

## 1.
$$ y'=y+x \\ y(0)=0 \\ x \in [0,2] $$

```python
def func_1(x, y):
    return y + x


args_1, vals_1 = diffeqs.euler(0, 2, func_1, 100, 0)
args_2, vals_2 = diffeqs.runge(0, 2, func_1, 100, 0)

plt.plot(args_1, vals_1)
plt.plot(args_2, vals_2)
plt.show()
```

## 2.
$$ y'=y-\frac{2x}{y} \\ y(0)=1 \\ x \in [0,40] $$

```python
def func_2(x, y):
    return y - 2 * x / y


args_1, vals_1 = diffeqs.euler(0, 40, func_2, 100, 1)
args_2, vals_2 = diffeqs.runge(0, 40, func_2, 100, 1)

plt.plot(args_1, vals_1)
plt.plot(args_2, vals_2)
plt.show()
```

## 3.
$$ y'=x^2-2y \\ y(0)=1 \\ x \in [0,10] $$

```python
def func_3(x, y):
    return x ** 2 - 2 * y


args_1, vals_1 = diffeqs.euler(0, 10, func_3, 100, 1)
args_2, vals_2 = diffeqs.runge(0, 10, func_3, 100, 1)

plt.plot(args_1, vals_1)
plt.plot(args_2, vals_2)
plt.show()
```
